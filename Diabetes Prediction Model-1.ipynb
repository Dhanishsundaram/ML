Importing the Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
# loading the diabetes dataset into pandas dataframe

db_data=pd.read_csv("/content/diabetes.csv")
```

```python
# printing the first five rows
db_data.head()
```

```python
## printing the total number of rows ,columns
db_data.shape
```

```python
#Describe the parameters of our data (Statistical definition)
db_data.describe()
```

```python
#Number of Diabetes and Non diabetes
db_data["Outcome"].value_counts()
```

0--->Non-Diabetes

1--->Diabetes


```python
db_data.groupby("Outcome").mean()
```

```python
#Separating the Data's and Labels
x=db_data.drop(columns="Outcome",axis=1)
y=db_data["Outcome"]
```

```python
print(x);
print(y);
```

Data Standardization

```python
scalar=StandardScaler()
```

```python
scalar.fit(x)
s=scalar.transform(x)
```

```python

print(s)
```

```python
x=s
y=db_data["Outcome"]
```

```python
print(x)
print(y)
```

Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the Model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#Training the svm classifier
classifier.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy score
x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy Score : ",training_data_accuracy)
```

```python
#Training the test data
x_test_prediction=classifier.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy Score : ",testing_data_accuracy)
```

Predictive System

```python
input_data=(10,168,74,0,0,38,0.537,34)

#Change  into numpy array
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)
#Standardize the input data
std_data=scalar.transform(input_data_reshape)
print(std_data)
prediction=classifier.predict(std_data)
print(prediction)

if(prediction[0]== 0):
  print("The Person has no Diabetes")
else:
  print("The person has Diabetes")
```
